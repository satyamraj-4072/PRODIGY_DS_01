# **Task -** Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

# **Coder - Satyam Raj**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset from Seaborn
titanic_data = sns.load_dataset("titanic")
```

```python
# Create a histogram to visualize the distribution of ages
plt.figure(figsize=(5, 5))
sns.histplot(data=titanic_data, x='age', bins=10, kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.tight_layout()
plt.show()
```

```python
# Create a bar chart to visualize the distribution of genders
plt.figure(figsize=(5, 5))
sns.countplot(data=titanic_data, x='sex')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
